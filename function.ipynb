```python
import json

def ask_for_input():
    print("*******   PHONEBOOK MAIN PAGE:   ******\n")
    print("******* add new contact(PRESS 1) ******")
    print("*******  update contact(PRESS 2) ******")
    print("******* look up contact(PRESS 3) ******")
    print("*******  delete contact(PRESS 4) ******")
    print("*******list all contacts(PRESS 5)******")
    print("*******      quit(PRESS 6)       ******")
    choice = input('\nplease enter your choice\n')
    return choice


def add_new_contact():
    name = input('enter the name - ')
    phone_number = input('enter the phone number - ')
    if check_if_input_is_empty(name) and validate_phone_number(phone_number):
        phonebook[name]=phone_number
        print('\nphone number successfully added!!')
    else:
        print('\nInvalid input, please try again')

def update_contact():
    name = input('enter the name - ')
    new_phone_number = input("new phone number - ")
    if check_if_input_is_empty(name) and validate_phone_number(new_phone_number):
        if name in phonebook:
            phonebook[name] = new_phone_number
            print(f'updated phone number for {name} is : {new_phone_number}')
        else:
            print('phone number not found ')
    else:
        print('Invalid input, please try again')
    
def search_contact():
    name = input('enter the name - ')
    if check_if_input_is_empty(name):
        if name in phonebook:
            phone_num = phonebook[name]
            print(f'phone number for {name} is : {phone_num}')
        else:
            print('phone number not found ')
    else:
        print('Invalid input, please try again')
        
def delete_contact():
    name = input('enter the name - ')
    if check_if_input_is_empty(name):
        if name in phonebook:
            del phonebook[name]
            print(f'successfully deleted {name} from phoneboook')
        else:
            print('phone number not found ')
    else:
        print('Invalid input, please try again')

def display_all_contacts():
    for name,number in phonebook.items():
        print(f'{name} - {number}')
        
def validate_phone_number(phone_num):
    if str(phone_num).isnumeric():
        if len(str(phone_num)) == 10:
            if str(phone_num).startswith('0'):
                return False
            else:
                return True
        else:
            return False
    else:
        return False  

def check_if_input_is_empty(input):
    if len(str(input).strip()) == 0:
        return False
    else:
        return True 

def save_phonebook_to_file():
    with open('phonebook.json', 'w') as json_file:
        json.dump(phonebook, json_file)

def load_phonebook_from_file():
    try:
        with open('phonebook.json', 'r') as json_file:
            return json.load(json_file)
    except FileNotFoundError:
        return {}

phonebook = load_phonebook_from_file()
count=0
while True:
    if count >=1:
        choice2=input(".....if you want to continue(PRESS 1)\nquit(PRESS 2)")
        if choice2=="1":
            choice = ask_for_input()
            if choice == '1':
                add_new_contact()
                save_phonebook_to_file()
            elif choice == '2':
                update_contact()
                save_phonebook_to_file()
            elif choice == '3':
                search_contact()
            elif choice == '4':
                delete_contact()
                save_phonebook_to_file()
            elif choice == '5':
                display_all_contacts()
            elif choice == '6':
                print("Exiting the phonebook application.")
                save_phonebook_to_file()
                break
            else:
                print("Invalid choice. Please choose again.")
        else:
            print("Exiting the phonebook application.")
            break
    else:
        count=count+1
        choice = ask_for_input()
        if choice == '1':
            add_new_contact()
            save_phonebook_to_file()
        elif choice == '2':
            update_contact()
            save_phonebook_to_file()
        elif choice == '3':
            search_contact()
        elif choice == '4':
            delete_contact()
            save_phonebook_to_file()
        elif choice == '5':
            display_all_contacts()
        elif choice == '6':
            print("Exiting the phonebook application.")
            save_phonebook_to_file()
            break
        else:
            print("Invalid choice. Please choose again.")
```
